```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.io import imread
import os
```

```python
leaf_data = pd.read_csv("train.csv", low_memory=False)
```

```python
leaf_names = dict()
unique_leaves = leaf_data["species"].unique()
for i in range(99):
    leaf_names[unique_leaves[i]] = i
for i in range(leaf_data.shape[0]):
    leaf_data.at[i,"species"] = leaf_names[leaf_data.at[i,"species"]]
```

```python
from skimage.transform import resize

image_dict = dict()
for f in os.listdir("./images/"):
    img = imread("./images/"+f)
    image_resized = resize(img, (60,60))
    image_dict[int(f[:-4])] = np.array(image_resized).flatten().tolist()
    #print(len(image_dict[int(f[:-4])]))
```

```python
leaf_data_for_pca = list(image_dict.values())
print(len(leaf_data_for_pca))
```

```python
from sklearn.decomposition import PCA
leaf_data_for_pca = np.array(leaf_data_for_pca)
pca = PCA(n_components=60)
leaf_data_after_pca = pca.fit_transform(leaf_data_for_pca)
#leaf_data_after_pca = pca.singular_values_
print(leaf_data_after_pca)
```

```python
print(len(leaf_data_after_pca))
```

```python
leaf_features = dict()
i = 0
for k in image_dict.keys():
    leaf_features[k] = leaf_data_after_pca[i]
    i+=1
```

```python
print(leaf_features)
```

```python
#Create new columns
for i in range(60):
    leaf_data["PCA"+str(i+1)] = 0.0
```

```python
leaf_data.columns
```

```python
leaf_data_extended = leaf_data.copy()
for i in range(leaf_data.shape[0]):
    for j in range(60):
        leaf_id = leaf_data_extended.at[i, "id"]
        leaf_data_extended.at[i, "PCA"+str(j+1)] = leaf_features[leaf_id][j]
```

```python
leaf_data_extended.describe().transpose()
```

```python
leaf_data_test = pd.read_csv("test.csv", low_memory=False)
```

```python
for i in range(60):
    leaf_data_test["PCA"+str(i+1)] = 0.0
leaf_data_test_extended = leaf_data_test.copy()
for i in range(leaf_data_test.shape[0]):
    for j in range(60):
        leaf_id = leaf_data_test_extended.at[i, "id"]
        leaf_data_test_extended.at[i, "PCA"+str(j+1)] = leaf_features[leaf_id][j]
```

```python
#Improving the baseline Random Forest model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

model_data = leaf_data_extended.copy()
rf_classifier = RandomForestClassifier(n_estimators=55, max_features=12)
train_Y = list(model_data["species"].values)
train_X = model_data.copy().loc[:, model_data.columns != "species"]
del train_X["id"]


#Creating the train-test splits
rf_train_X, rf_test_X, rf_train_Y, rf_test_Y = train_test_split(train_X, train_Y, test_size=0.2, random_state=23)
rf_classifier.fit(rf_train_X, rf_train_Y)
prob_values = rf_classifier.predict_proba(rf_test_X)
values = rf_classifier.predict(rf_test_X)
print("Accuracy from direct prediction: ", accuracy_score(rf_test_Y,values))

#Computing the accuracy of the model

acc = 0.0
for i in range(len(rf_test_Y)):
    max_index = np.where(prob_values[i] == prob_values[i].max())
    if rf_test_Y[i] in max_index[0].tolist():
        acc += 1.0
print("Accuracy from probability prediction: ", acc/len(rf_test_Y))
```

```python
final_test_X = leaf_data_test.copy()
del final_test_X["id"]
```

```python
f = open("sample_submission.csv","r")
submission_columns = f.readline().strip().split(",")
f.close()
```

```python
id_values = leaf_data_test["id"]
final_list = list()
for i in range(leaf_data_test.shape[0]):
    tmp = [id_values[i]]
    tmp.extend(final_test_results[i])
    final_list.append(tmp)
submission_df = pd.DataFrame(final_list ,columns=submission_columns)
submission_df.to_csv('random_forest_improved.csv', index=False, float_format='%.6f')
```
